```python
import numpy as np
import pandas as pd
```

```python
data_csvfile = pd.read_csv('Berlin-tram-route - Sheet1.csv')
```

```python
New_column_name = list(data_csvfile.columns[1:]) 
while len(New_column_name)>0:
  cluster = [New_column_name[0]]
  New_column_name.remove(cluster[0])
  k = New_column_name.copy()
  for i in k:
    k =0
    for j in cluster:
      if data_csvfile[data_csvfile["route"]==j][i].values[0]==0:
        k = k+1
    if k == len(cluster):
      cluster.append(i)
      New_column_name.remove(i)
  print(cluster) 
```
